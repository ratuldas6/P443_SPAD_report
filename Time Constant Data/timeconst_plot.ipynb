```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
filename = 'dat_150k.txt'
counts = np.loadtxt(filename, delimiter=',', skiprows=1, dtype=int)

n, bins, patches = plt.hist(counts, 30, alpha=0.75)

plt.xlabel('Length of counts')
plt.ylabel('Counts')
plt.title('Counts vs Time between counts')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
#plt.xlim(40, 160)
#plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
```
