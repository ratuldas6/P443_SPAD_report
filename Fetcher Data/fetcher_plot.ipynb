```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
data = pd.read_csv('Data_trial6_dark.csv')
counts = data['p_det']

#n, bins, patches = plt.hist(counts, 10, density=True, facecolor='g', alpha=0.75)
n, bins, patches = plt.hist(counts, 8, facecolor='g', alpha=0.75)

plt.xlabel('Counts')
plt.ylabel('Frequency of counts')
plt.title('Histogram of 20ms data in darkness')
#plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
#plt.xlim(40, 160)
#plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
```
